## Polynomial Regression 

#### Used for non linear data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline

%matplotlib inline
```

```python
# Generate random data points for a quadratic equation y = ax^2 + bx + c
np.random.seed(42)  # for reproducibility
n_points = 100
X = np.linspace(-10, 10, n_points)
a, b, c = 1, 2, 3  # Coefficients for the quadratic equation

# Add some noise to make it look more realistic
noise = np.random.normal(0, 5, size=n_points)
y = a * X**2 + b * X + c + noise 
X = X.reshape(-1,1)
```

```python
X,y
```

```python
## splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Linear Regression

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
y_pred = regression.predict(X_test)

print("r2_score :", r2_score(y_test, y_pred))

plt.scatter(X_train, y_train, color='blue', label = "Train data points")
plt.plot(X_train, regression.predict(X_train), color='r')
plt.xlabel("X train data")
plt.ylabel("y train data")
plt.legend()
plt.show()

plt.scatter(X_test, y_test, color='green', label = "Test data points")
plt.plot(X_test, regression.predict(X_test), color='r')
plt.xlabel("X test data")
plt.ylabel("y test data")
plt.legend()
plt.show()
```

## Polynomial regression

```python
# Step 1: Generate polynomial features (degree=2, include_bias=True means it includes the bias term or intercept)
poly = PolynomialFeatures(degree=2, include_bias=True)

# Step 2: Transform the original training data (X_train) into polynomial features
X_train_poly = poly.fit_transform(X_train)

# Step 3: Transform the original test data (X_test) into polynomial features, using the same transformation
X_test_poly = poly.transform(X_test)

# Step 4: Fit the regression model using the polynomial-transformed training data
regression.fit(X_train_poly, y_train)

# Step 5: Predict the target variable (y) for the test data using the polynomial-transformed test data
y_pred_poly = regression.predict(X_test_poly)

# Step 6: Evaluate the model performance using the R^2 score for test data
print("r2_score :", r2_score(y_test, y_pred_poly))

# Step 7: Plot the training data points and the predicted points for the training data
plt.scatter(X_train, y_train, color='blue', label="Train data points")  # Actual training data points
plt.scatter(X_train, regression.predict(X_train_poly), color='r', label="Predicted points")  # Predicted points
plt.xlabel("X train data")  # X-axis label for training data plot
plt.ylabel("y train data")  # Y-axis label for training data plot
plt.legend()  # Display the legend
plt.show()  # Show the plot

# Step 8: Plot the test data points and the predicted points for the test data
plt.scatter(X_test, y_test, color='green', label="Test data points")  # Actual test data points
plt.scatter(X_test, regression.predict(X_test_poly), color='r', label="Predicted points")  # Predicted points
plt.xlabel("X test data")  # X-axis label for test data plot
plt.ylabel("y test data")  # Y-axis label for test data plot
plt.legend()  # Display the legend
plt.show()  # Show the plot
```

```python
# Step 1: Generate polynomial features (degree=3, include_bias=True means it includes the bias term or intercept)
poly = PolynomialFeatures(degree=3, include_bias=True)

# Step 2: Transform the original training data (X_train) into polynomial features
X_train_poly = poly.fit_transform(X_train)

# Step 3: Transform the original test data (X_test) into polynomial features, using the same transformation
X_test_poly = poly.transform(X_test)

# Step 4: Fit the regression model using the polynomial-transformed training data
regression.fit(X_train_poly, y_train)

# Step 5: Predict the target variable (y) for the test data using the polynomial-transformed test data
y_pred_poly = regression.predict(X_test_poly)

# Step 6: Evaluate the model performance using the R^2 score for test data
print("r2_score :", r2_score(y_test, y_pred_poly))

# Step 7: Plot the training data points and the predicted points for the training data
plt.scatter(X_train, y_train, color='blue', label="Train data points")  # Actual training data points
plt.scatter(X_train, regression.predict(X_train_poly), color='r', label="Predicted points")  # Predicted points
plt.xlabel("X train data")  # X-axis label for training data plot
plt.ylabel("y train data")  # Y-axis label for training data plot
plt.legend()  # Display the legend
plt.show()  # Show the plot

# Step 8: Plot the test data points and the predicted points for the test data
plt.scatter(X_test, y_test, color='green', label="Test data points")  # Actual test data points
plt.scatter(X_test, regression.predict(X_test_poly), color='r', label="Predicted points")  # Predicted points
plt.xlabel("X test data")  # X-axis label for test data plot
plt.ylabel("y test data")  # Y-axis label for test data plot
plt.legend()  # Display the legend
plt.show()  # Show the plot
```

## polynomial pipeline

```python
def poly_regression(degree):

    X_new = np.linspace(-15, 15, 30).reshape(-1,1)
    # X_new = np.linspace(-3, 3, 200).reshape(200, 1)

    poly_features = PolynomialFeatures(degree=degree, include_bias=True)
    lin_regression = LinearRegression()

    poly_regression = Pipeline([

        ('poly_features', poly_features),
         ('lin_regression', lin_regression)
    ])

    poly_regression.fit(X_train, y_train)

    y_new_pred = poly_regression.predict(X_new)

    plt.scatter(X_train, y_train, color = 'b')
    plt.scatter(X_test, y_test, color='g')
    plt.plot(X_new, y_new_pred, color = 'r', label = "degree" + str(degree))
    plt.xlabel('X')
    plt.ylabel('y')
    plt.axis([-15,15, 0, 150])
    plt.legend()
    plt.show()
```

```python
poly_regression(1)  # with degree of 1
```

```python
poly_regression(3)  # with degree of 3
```
